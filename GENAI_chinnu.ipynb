```python
from transformers import GPT2Tokenizer,GPT2LMHeadModel
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
```

```python
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tokenizer.eos_token_id)
```

```python
prompt = "I loves Travelling"
numeric_id = tokenizer.encode(prompt, return_tensors = 'pt')
```

```python
numeric_id
```

```python
result = model.generate(numeric_id, max_length = 100, num_beams=5,no_repeat_ngram_size=2, early_stopping=True)
```

```python
generated_text = tokenizer.decode(result[0], skip_special_tokens=True)
print(generated_text)
```

Coding: Summarize text using a pre-trained model

```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """Artificial Intelligence (AI) is revolutionizing the way we interact with technology.
It is used in various fields such as healthcare, finance, and education.
AI-powered chatbots and virtual assistants have made customer support more efficient.
Additionally, AI is helping researchers analyze large datasets quickly, leading to faster discoveries and innovations."""
```

```python
summary = summarizer(text, max_length=50, min_length=20, do_sample=False)
```

```python
print(summary)
```
